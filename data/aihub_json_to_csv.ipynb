```python
import pandas as pd
import json
import os
```

```python
path_T = 'Training/02.라벨링데이터'

folder_list = os.listdir(path_T)

json_file_list_T = []

for folder in folder_list:
    folder_path = os.path.join(path_T, folder)

    if os.path.isdir(folder_path):
        files_in_folder = os.listdir(folder_path)
        
        # 필요한 경우 파일을 추가로 필터링
        json_files_in_folder = [file for file in files_in_folder if file.endswith('.json')]

        # 각 JSON 파일의 전체 경로를 json_file_list에 추가
        json_file_list_T.extend([os.path.join(folder_path, json_file) for json_file in json_files_in_folder])
```

```python
dfs = []

# 목록의 각 JSON 파일을 반복
for json_file_path in json_file_list_T:
    with open(json_file_path) as f:
        js = json.loads(f.read())

    # 현재 JSON 파일의 DataFrame 생성
    df = pd.DataFrame(js, columns=["Index", "RawText", "ReviewScore", "ProductName"])

    # DataFrame을 DataFrame 목록에 추가
    dfs.append(df)

# 목록의 모든 데이터 프레임을 하나의 데이터 프레임으로 연결
combined_df_T = pd.concat(dfs, ignore_index=True)
```

```python
combined_df_T
```

```python
combined_df_T.to_csv('train.csv', index=False)
```

```python
path_V = 'Validation/02.라벨링데이터'

folder_list = os.listdir(path_V)

json_file_list_V = []

for folder in folder_list:
    folder_path = os.path.join(path_V, folder)

    if os.path.isdir(folder_path):
        files_in_folder = os.listdir(folder_path)
        
        # 필요한 경우 파일을 추가로 필터링
        json_files_in_folder = [file for file in files_in_folder if file.endswith('.json')]

        # 각 JSON 파일의 전체 경로를 json_file_list에 추가
        json_file_list_V.extend([os.path.join(folder_path, json_file) for json_file in json_files_in_folder])

# json_file_list는 하위 디렉토리에 있는 모든 JSON 파일의 전체 경로를 포함
```

```python
dfs = []

# 목록의 각 JSON 파일을 반복
for json_file_path in json_file_list_V:
    with open(json_file_path) as f:
        js = json.loads(f.read())

    # 현재 JSON 파일의 DataFrame 생성
    df = pd.DataFrame(js, columns=["Index", "RawText", "ReviewScore", "ProductName"])

    # DataFrame을 DataFrame 목록에 추가
    dfs.append(df)

# 목록의 모든 데이터 프레임을 하나의 데이터 프레임으로 연결
combined_df_V = pd.concat(dfs, ignore_index=True)
```
